```python
m = df['Vaccinated Rate'].mean()
md = df['Vaccinated Rate'].median()
sd = df['Vaccinated Rate'].std()
print("Mean Vaccinated Rate :",m)
print("Median Vaccinated Rate :",md)
print("Standard Deviation Vaccinated Rate :",sd)
```

```python
Q1 = df['Vaccinated Rate'].quantile(0.25)
Q3 = df['Vaccinated Rate'].quantile(0.75)
IQR = Q3 - Q1
print("IQR Vaccinated Rate :",IQR)
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
print("Lower Limit Vaccinated Rate :",lower_limit)
print("Upper Limit Vaccinated Rate :",upper_limit)
```
